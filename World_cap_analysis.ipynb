```python
matches=[]#creates an empty list to store the matches from the world cup data
for round_ in world_cup_data["rounds"]:#loops through the rounds in the world cup data and extracts the matches for each round
    round_matches=round_["matches"]#create a new list of matches extracted with the for loop to extend to the matches list
    matches.extend(round_matches)#extend is used to add the matches from each round to the matches list
#round_names=[round_["name"] for round_ in world_cup_data["rounds"]]#creates a list of round names using a list comprehension that iterates through the rounds in the world cup data and extracts the name of each round

matches[0]#prints the first match in the matches list, which is a dictionary containing information about the match such as the teams playing, the score, and the date of the match
```

```python
teams_set=set()#creates an empty set to store the unique team names from the matches in the world cup data
for match in matches:#loops through the matches in the matches list and prints the home team, away team, and score for each match
    teams_set.add(match['team1']['name'])
    teams_set.add(match['team2']['name'])

    teams=sorted(list(teams_set))#creates a sorted list of team names from the teams set
```

```python
def find_winner(match):
    if match["score1"] > match["score2"]:
        return match["team1"]["name"]
    elif match["score2"] > match["score1"]:
        return match["team2"]["name"]
    else:
        return None
```

```python
assert find_winner(matches[0]) == "Russia"
assert find_winner(matches[1]) == "Uruguay"
assert find_winner(matches[2]) == None
```

```python
def normalized_location(country_name):
    name_sub_dict = {
        "Russian Federation": "Russia",
        "Egypt, Arab Rep.": "Egypt",
        "Iran, Islamic Rep.": "Iran",
        "Korea, Rep.": "South Korea",
        "United Kingdom": "England"
    }
    return name_sub_dict.get(country_name, country_name)
print(normalized_location("Russian Federation"))
```

```python
population_data_filtered = []
for record in population_data:
    normalized_name= normalized_location(record["Country Name"])

    if normalized_name in teams and record["Year"] == "2018":
        record["Country Name"]=normalized_name
        population_data_filtered.append(record)
len(population_data_filtered)
```
